The options sensitivities (Greeks), are the partial derivatives of the BSM formula. They measure the sensitivty of the option price to a small change in a parameter of the formula.

### 2.1 Delta Greeks

### 2.1.1 Delta

Delta is the option's sensitivity to small change in the underlying asset price

#### Call

$$\Delta_{call} = \frac{\delta c}{\delta S} = e^{\left( b - r \right) T} N \left( d_1 \right) \gt 0$$


#### Put

$$\Delta_{put} = \frac{\delta p}{\delta S} = e^{\left( b - r \right) T} [ N \left( d_1 \right) - 1 ] \lt 0$$
